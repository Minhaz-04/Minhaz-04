```python
plt.figure(figsize=(8,6))
plt.bar(df['age'],df['income'],color='green')
plt.xlabel('age')
plt.ylabel('income')
plt.title('age vs income')
plt.grid(axis='y')
```

```python
plt.figure(figsize=(8,6))
plt.bar(df['year'],df['expenditure'],color='blue')
plt.xlabel('year')
plt.ylabel('expenditure')
plt.title('year vs expenditure')
plt.grid(axis='y')
```

```python
plt.figure(figsize=(8,6))
plt.plot(df['year'],df['eqv_exp'],marker='o',color='red',label='eqv_exp')
plt.xlabel('year')
plt.ylabel('eqv_exp')
plt.title('year vs eqv_exp')
plt.grid(axis='y')
```

```python
plt.figure(figsize=(8,6))
plt.plot(df['year'],df['income'],marker='o',color='green',label='income')
plt.xlabel('year')
plt.ylabel('income')
plt.title('year vs income')
plt.grid(axis='y')
```
